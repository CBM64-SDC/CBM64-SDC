```python
import keras
import cv2
import numpy as np

from keras.layers.advanced_activations import LeakyReLU
from keras.models import Model, Sequential
from keras.optimizers import Adam
from keras.layers import MaxPooling2D, Dense, Flatten, Reshape, GlobalAveragePooling2D, Conv2D, Dropout

import matplotlib.pyplot as plt
import matplotlib

from darkflow.net.build import TFNet
import time

from glob import glob


%matplotlib inline
```

```python
## The model implemented below is the tiny yolo architecture taken from the yolo research paper.
## since training requires a huge computing powers, and based on the advice taken from many experts i decided 
## to use the pretrained weights provided by the yolo teams themselves, since finding a cars dataset and annotating
## each picture is a completely time consuming process

model = Sequential()

model.add(Conv2D(16, (3, 3), padding = 'same', input_shape=(416,416,3)))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2, 2), padding = 'valid'))

model.add(Conv2D(64, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2, 2), padding = 'valid'))

model.add(Conv2D(128, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2,2), padding = 'valid'))

model.add(Conv2D(128, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2,2), padding = 'valid'))

model.add(Conv2D(256, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2,2), padding = 'valid'))

model.add(Conv2D(512, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))
model.add(MaxPooling2D(pool_size=(2,2), padding = 'valid'))

model.add(Conv2D(1024, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))

model.add(Conv2D(1024, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))

model.add(Conv2D(1024, (3, 3), padding = 'same'))
model.add(LeakyReLU(0.1))

model.add(Flatten())

model.add(Dense(256))
model.add(Dense(4096))

model.add(LeakyReLU(0.1))

model.add(Dense(1470))

model.summary()
```

```python
args = {
    'model': 'cfg/tiny-yolo-voc.cfg', ## using darknet's implemented annotations
    'load': 'bin/tiny-yolo-voc.weights', ## using darknet's pretrained weights
    'treashold': 0.15
}

yolo = TFNet(args) 
```

```python
image = cv2.imread('./car-pictures/car-gt.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
car_pictures = glob('./car-pictures/*.jpg')


for picture in car_pictures:
    start_time = time.time() ## will be used later for calculating predicting time for each picture
    image = cv2.imread(picture)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    results = yolo.return_predict(image)
    for result in results:
        if result['label'] == 'car':
            print(result)
            top_left = (result['topleft']['x'], result['topleft']['y'])
            bottom_right = (result['bottomright']['x'], result['bottomright']['y'])
            image = cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 4)
    plt.imshow(image)
    plt.show()
    print('FPS {:.1f}'.format(1 / (time.time() - start_time)))
    print("\n")
```
